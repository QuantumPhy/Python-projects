```python
import sys


key_value = dict()
for filename in ['raven_01.txt', 'raven_02.txt', 'raven_03.txt']:
    if(filename != sys.argv[0]):
        in_file = open(filename, 'r')
        key_value_per_file = my_map(in_file)
        for key in key_value_per_file:
            if(key in key_value):
                key_value[key].append(key_value_per_file[key])
            else:
                key_value[key]=[key_value_per_file[key]]
reduced = dict()
for key in key_value:
    reduced[key] = my_reduce(key_value[key])
for key in reduced:
    print(key + ": " + str(reduced[key]))
```

```python
import re
def my_map(in_file):
    wc = dict()
    for line in in_file:
        line = re.sub("[^A-Za-z0-9]"," ",line)
        words = line.split()
        for word in words:
            if(word in wc):
                wc[word]+=1
            else:
                wc[word]=1
    return wc
```

```python
def my_reduce(value):
    number = 0
    for v in value:
        number+=v
    return number
```
